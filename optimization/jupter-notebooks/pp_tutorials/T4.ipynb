## Collection functions

visualize each line longer than 2km

```python
# YouTube Tutorial: https://www.youtube.com/watch?v=4LQLZQWPaPM
import pandapower.networks as nw
import pandapower.plotting as plot
import matplotlib.pyplot as plt
import seaborn as sns
import colorcet as cc

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a palette with 180 colors using a cyclic colormap
colors = sns.color_palette("hsv", 179)
print(colors[0])
# Show the palette
sns.palplot(colors)
plt.show()
```

```python
# Load the network
net = nw.mv_oberrhein()

# Create buses collections
bc = plot.create_bus_collection(net, buses=net.bus.index, color='red', size=80, zorder=1)

# Create lines collections
lc = plot.create_line_collection(net, lines=net.line.index, color='gray', zorder=2)
```

```python
plot.draw_collections([lc, bc])
plt.show()
```

```python
long_lines = net.line.loc[net.line.length_km > 2.].index
long_lines
```

```python
lcl = plot.create_line_collection(net, lines=long_lines, color='green', zorder=2)
plot.draw_collections([lc, bc, lcl])
plt.show()
```
